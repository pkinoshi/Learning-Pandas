#### Inner Vs Outer joins

###### the how parameter

```python
ivies.merge(regions)# same as pd.merge(ivies, regions)
```

```python
pd.merge(ivies, regions, how='inner')# the default value for the how parameter is 'inner' same as intersection of set
```

```python
pd.merge(ivies, regions, how='outer')#same as union of set
```

```python
pd.merge(ivies, regions, how='outer')['Starting Median Salary'].hasnans
```

```python
np.count_nonzero(pd.merge(ivies, regions, how='outer')['Starting Median Salary'].isna())
```

```python
a = {1,2,3}
b = {4,5,6}
```

```python
c = {1,3,4,6}
```

```python
a.union(b)
```

```python
a.intersection(b)
```

```python
a.intersection(c)
```

```python
b.intersection(c)
```

#### Left Vs Right joins

```python
pd.merge(ivies, regions, how='left')
```

```python
pd.merge(ivies, regions, how='right')
```

```python
pd.merge(ivies, regions, how='right').equals(pd.merge(ivies, regions, how='outer'))
```

###### From the above line of code, it is safe to say that "how='outer'" is the same as "how='right'"
