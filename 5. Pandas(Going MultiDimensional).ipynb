##### Build Manually

###### criteria:<br>-2 level multiindex column axis, containing:<br>- trading volume(high and low) and tickers(MSFT & AMZN)<br>- 10 random closing price observations

```python
tech.reset_index(inplace=True)
```

```python
tech.head()
```

```python
#Step 1: Create multiindex
pd.MultiIndex.from_product([['low', 'high'], ['MSFT', 'AMZN']],
                            names=['Volume Category', 'Ticker'])
```

```python
cols = pd.MultiIndex.from_product([['low', 'high'], ['MSFT', 'AMZN']],
                            names=['Volume Category', 'Ticker'])
```

```python
cols
```

```python
#Step 2: prep the underlying data
low = tech['Volume Category'] =='low'
high= tech['Volume Category'] =='high'
amzn = tech.Ticker == 'AMZN'
msft = tech.Ticker == 'MSFT'
```

```python
tech[tech['Volume Category'] == 'low']
```

```python
tech.head()
```

```python
tech.loc[(tech['Volume Category']=='low') & (tech['Ticker']=='AMZN')]
```

```python
tech.loc[(tech['Volume Category']=='low') & (tech['Ticker']=='AMZN')].close.sample(10).values
```

```python
tech.loc[low & amzn].close.sample(10).values
```

```python
tech.loc[low & msft].close.sample(10).values
```

```python
data = [
        tech[low & msft].close.sample(10).values,
        tech[low & amzn].close.sample(10).values,
        tech[high & msft].close.sample(10).values,
        tech[high & amzn].close.sample(10).values
       ]
```

```python
tech[low & msft].close.sample(10).values
```

```python
data = [ [*i] for i in zip(*data)]
```

```python
data
```

```python
df = pd.DataFrame(data, columns=cols)
```
